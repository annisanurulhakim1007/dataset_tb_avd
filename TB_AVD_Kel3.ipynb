```python
# import packages
import pandas as pd
import numpy as np
```

```python
# membaca dataset
df = pd.read_csv('https://raw.githubusercontent.com/annisanurulhakim1007/dataset_tb_avd/main/covid_19_indonesia_time_series_all.csv')

# pengaturan display untuk menampilkan semua kolom
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
# melihat dimensi data
df.shape
```

```python
# melihat tipe data setiap kolom pada dataset
df.dtypes
```

```python
# menghitung jumlah data
df.count()
```

```python
# menghapus kolom yang tidak sesuai
df = df.drop(['Location ISO Code', 'City or Regency', 'Province', 'Country', 'Continent', 'Time Zone', 'Special Status'], axis=1)

# pengaturan display untuk menampilkan semua kolom
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
# mengganti nilai NaN di kolom 'Island' dengan 'Indonesia' jika 'Location Level' adalah 'Country'
df.loc[df['Location Level'] == 'Country', 'Island'] = 'Indonesia'

# mengganti nilai NaN di kolom 'Total Rural Villages' dengan 0 jika 'Location' adalah 'DKI Jakarta'
df.loc[df['Location'] == 'DKI Jakarta', 'Total Rural Villages'] = 0

# pengaturan display untuk menampilkan semua kolom
pd.set_option('display.max_columns', None)
df.head(10)
```

```python
# mengubah kolom 'Date' yang masih bertipe Object menjadi tipe Date
df['Date'] = pd.to_datetime(df['Date'])

# pengaturan display untuk menampilkan semua kolom
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
# melihat tipe data setiap kolom pada dataset
df.dtypes
```

```python
# melihat jumlah data yang duplikat
duplicate_rows_df = df[df.duplicated()]
print("Jumlah data duplikat: " , duplicate_rows_df.shape)
```

```python
# menampilkan jumlah data yang kosong
print(df.isnull().sum())
```

```python
# menghapus baris data yang kosong
df = df.dropna()

# menampilkan jumlah data yang kosong
print(df.isnull().sum())
```

```python
# melihat dimensi data
df.shape
```

```python
# menghitung jumlah data
df.count()
```

```python
# membuat tabel rata-rata kasus baru per provinsi
kasus_baru = df[df['Location Level'] == 'Province'].groupby('Location')['New Cases'].mean()

# mengubah Series menjadi DataFrame
kasus_baru = kasus_baru.reset_index()

# mengubah nama kolom
kasus_baru.columns = ['Location', 'Average New Cases']

# ambil data Longitude dan Latitude dari DataFrame asli
df_location = df[['Location', 'Longitude', 'Latitude']].drop_duplicates()

# gabungkan 'average_new_cases' dengan 'df_location'
kasus_baru = pd.merge(kasus_baru, df_location, on='Location', how='left')

# urutkan kolom
kasus_baru = kasus_baru.reindex(columns=['Location', 'Longitude', 'Latitude', 'Average New Cases'])
kasus_baru
```

```python
# import package
import folium

# buat peta dasar
m = folium.Map(location=[-2.548926, 118.0148634], zoom_start=5)  # koordinat pusat Indonesia

# tambahkan titik data
for idx, row in kasus_baru.iterrows():
    # buat popup dengan informasi rata-rata kasus baru
    popup = folium.Popup('Average New Cases: ' + str(row['Average New Cases']), max_width=300)

    # tambahkan marker ke peta
    folium.Marker([row['Latitude'], row['Longitude']], popup=popup).add_to(m)

# tampilkan peta grafik
m
```

```python
# membuat tabel rata-rata pemulihan per provinsi
kematian_pemulihan = df[df['Location Level'] == 'Province'].groupby('Location')[['New Deaths', 'New Recovered']].mean()

# mengubah Series menjadi DataFrame
kematian_pemulihan = kematian_pemulihan.reset_index()

# mengubah nama kolom
kematian_pemulihan.columns = ['Location',  'Average New Deaths', 'Average New Recovered']

# ambil data Longitude dan Latitude dari DataFrame asli
df_location = df[['Location', 'Longitude', 'Latitude']].drop_duplicates()

# gabungkan 'average_new_cases' dengan 'df_location'
kematian_pemulihan = pd.merge(kematian_pemulihan, df_location, on='Location', how='left')

# urutkan kolom
kematian_pemulihan = kematian_pemulihan.reindex(columns=['Location', 'Longitude', 'Latitude', 'Average New Deaths', 'Average New Recovered'])
kematian_pemulihan
```

```python
# buat peta dasar
m = folium.Map(location=[-2.548926, 118.0148634], zoom_start=5)  # koordinat pusat Indonesia

# tambahkan titik data untuk rata-rata pemulihan dan kematian
for idx, row in kematian_pemulihan.iterrows():
    # buat informasi popup
    popup_html = f'<b>Provinsi:</b> {row["Location"]}<br>'
    popup_html += f'<b>Average New Deaths:</b> {row["Average New Deaths"]}<br>'
    popup_html += f'<b>Average New Recovered:</b> {row["Average New Recovered"]}<br>'

    # tambahkan CircleMarker ke peta dengan popup yang berisi informasi rata-rata kematian dan pemulihan
    folium.CircleMarker(location=[row['Latitude'], row['Longitude']],
                        radius=5,
                        popup=folium.Popup(popup_html, max_width=300),
                        color='red',
                        fill=True,
                        fill_color='red').add_to(m)

# tampilkan peta grafik
m
```
